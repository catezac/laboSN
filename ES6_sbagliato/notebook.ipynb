```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

magn = np.loadtxt("NSL_SIMULATOR/OUTPUT/magnetization.dat")
x = magn[:,0]
y = magn[:,2]
err = magn[:,3]


plt.figure(1)
plt.errorbar(x,y,err, ecolor= 'orange' ,color='red')
plt.xlabel('block')
plt.ylabel('magnetization')
plt.grid(True)
```
