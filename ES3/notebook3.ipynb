```python
#call
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

Call = np.loadtxt("Call.dat")
x = Call[:,0]
y = Call[:,1]
err = Call[:,2]

plt.figure(1)
plt.errorbar(x,y,err, ecolor= 'orange' ,color='red', label = 'direct sampling')
plt.axhline(y=14.975790778311286, color='firebrick', linestyle='-')
plt.xlabel('#throws')
plt.ylabel('Call')
plt.grid(True)

Call_d = np.loadtxt("Call_d.dat")
x_d = Call_d[:,0]
y_d = Call_d[:,1]
err_d = Call_d[:,2]

plt.figure(1)
plt.errorbar(x_d,y_d,err_d, ecolor= 'darkcyan' ,color='blue', label = 'discrete sampling')
plt.axhline(y=14.975790778311286, color='firebrick', linestyle='-')
plt.xlabel('#throws')
plt.ylabel('Call')
plt.title("Call-option price")
plt.grid(True)
plt.legend()

plt.figure(2)
plt.plot(x_d, err_d, color = 'darkcyan', label = 'discrete sampling')
plt.plot(x, err, color = 'orange', label = 'direct sampling')
plt.grid()
plt.xlabel('#throws')
plt.ylabel('error')
plt.title("Call-option price error")


plt.legend()
```
